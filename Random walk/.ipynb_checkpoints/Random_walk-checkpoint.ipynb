## Random walk

Random walk is basically a simulation process in which movement of some object is determined probabilistically. For a simplistic model, we will look at the diffusion along one dimension (X-axis). However, we can extend this same logic for all dimensions. Here, diffusion depends on Brownian motion, which can be described as a random walk. This is how we can think of random walk as: You can think of a one-dimensional random walk in terms of coin flipping. Suppose you are standing on a very long straight sidewalk that extends both in front of and behind you. You flip a coin. If it comes up heads, you take a step forward; tails means to take a step backward.

Consider first a random walk in one dimension. Imagine a ‘walker’ (a particle) that takes one step during each unit time interval. Each step is one unit long, but the step direction is randomly chosen to be either ‘forward’ in the +x or ‘backward’ in the −x direction. In the codes written below, I have assigned the different conditions for this movement. Please check the condition for each one, and comment which condition I have assigned for each type of motion. For example, in the first code, I have subtracted 1 from the previous position of the particle, if the probability (which I genereated from the random numner) is less than 0.5. And added 1 from the previous position, if the probability os greated than 0.5. Again, these probabilities are not deterministic, as they are being generated randomly. So, the motion in one direction has no influence whatsover for the motion in one direction.

```python
import random as random
import matplotlib.pyplot as plt

class Randomwalk():
  def __init__(self,N,step,x0):
        self.N = N
        self.X = []
        self.step = step
        self.x0 = x0

  def walk(self):
    x = self.x0
    k = 0
    while k < self.N:
        x = x + self.step*random.choice([-1,1])
        self.X.append(x)
        k = k+1
    return self.X

  def view(self,label):
    plt.style.use('seaborn')
    plt.title('Randomwalk1D')
    plt.xlabel('Time')
    plt.ylabel('distance')
    plt.plot(self.walk(), label = label)
    plt.legend()
```

```python
A = Randomwalk(1000,5,0,)
B = Randomwalk(1000,3,0)
C = Randomwalk(1000,1,0)
```

```python
A.view('A')
B.view('B')
C.view('C')
```

## In two dimensions

```python
def func1(x1,y1):
    '''This function moves one step on randomly selected 
    direction'''
    import random as random
    Axis = random.choice(['x','y'])
    if Axis == 'x':
        a = random.choice([-1,1])
        if a == 1:
            x2 = x1+1
        else:
            x2 = x1-1
            
        y2 = y1
        
    else:
        a = random.choice([-1,1])
        if a == 1:
            y2 = y1+1
        else:
            y2 = y1-1
            
        x2 = x1
    return x2,y2

def func2(N,position):
    '''It will move N steps in 2d'''
    X = []
    Y = []
    '''starting positions'''
    xi=  position[0]
    yi = position[1]
    i = 0
    while i < N:
        x,y = func1(xi,yi)
        X.append(x)
        Y.append(y)
        '''updating position'''
        xi = x
        yi = y
        
        i += 1
    return X,Y


 
 
```

```python
import matplotlib.pyplot as plt
N = 1000
A = func2(N,[0,0])
plt.style.use('seaborn')
plt.figure(figsize = [15,10])
plt.plot(A[0],A[1],"-",color = 'red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Brownian Motion")
plt.show()
```

```python
def func1(x1,y1,z1):
    '''This function moves one step on randomly selected 
    direction'''
    import random as random
    Axis = random.choice(['x','y','z'])
    if Axis == 'x':
        a = random.choice([-1,1])
        if a == 1:
            x2 = x1+1
        else:
            x2 = x1-1
            
        y2 = y1
        z2 = z1
    if Axis == 'y':
        a = random.choice([-1,1])
        if a == 1:
            y2 = y1+1
        else:
            y2 = y1-1
            
        x2 = x1
        z2 = z1
    

        
    if Axis == 'z':
        a = random.choice([-1,1])
        if a == 1:
            z2 = z1+5
        else:
            z2 = z1-5
            
        x2 = x1
        y2 = y1
    return x2,y2,z2

def func2(N,position):
    '''It will move N steps in 3d'''
    X = []
    Y = []
    Z = []
    '''starting positions'''
    xi=  position[0]
    yi = position[1]
    zi = position[2]
    i = 0
    while i < N:
        x,y,z = func1(xi,yi,zi)
        X.append(x)
        Y.append(y)
        Z.append(z)
        '''updating position'''
        xi = x
        yi = y
        zi = z
        i += 1
    return X,Y,Z
```

```python
N = 1000
A = func2(N,[0,0,0])

fig = plt.figure(figsize = [15,10])
ax = plt.axes(projection='3d')
zdata = A[0]
xdata = A[1]
ydata = A[2]

plt.style.use('seaborn')
plt.plot(A[0],A[1],A[2],"-",color = 'red',linestyle = '', marker = '.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.ylabel("Z-axis")
plt.title("3D Brownian Motion")
plt.show()
```

## Theory versus simulation
To account for the right size for the fluctuations, the random force must be uncorrelated Gaussian noise with zero mean and variance  $$\sigma^2 = \frac{2 k_B T \Delta t}{m \gamma}$$

```python
#Parameters for the Brome Mosaic Virus
m = 7.6e-21      # kg
gamma = 3.4e10   # s^-1
dt = 10e-9       # s
T = 300          # K

kB = 1.380649e-23 # J/K

max_steps = 10000
x0 = 0           # m

sigma = np.sqrt(2*kB*T * dt / (m * gamma))  # m 
```

```python
#Parameters for the Brome Mosaic Virus
m = 7.6e-21      # kg
gamma = 3.4e10   # s^-1
dt = 10e-9       # s
T = 300          # K

kB = 1.380649e-23 # J/K

max_steps = 10000
x0 = 0           # m

sigma = np.sqrt(2*kB*T * dt / (m * gamma))  # m 
```

```python
rng = np.random.default_rng()
steps = rng.normal(loc=0, scale=sigma, size=max_steps)
x_trajectory = x0 + np.cumsum(steps)
```

```python
times = dt * np.arange(len(x_trajectory))


plt.plot(times, x_trajectory)
plt.xlabel("time (s)")
plt.ylabel("position (m)");

```

```python
N_repeats = 100
steps = rng.normal(loc=0, scale=sigma, size=(N_repeats, max_steps))


x_trajectories = x0 + np.cumsum(steps, axis=1)

times = dt * np.arange(x_trajectories.shape[1])
plt.plot(times, x_trajectories.T, alpha=0.5);
plt.xlabel("time (s)")
plt.ylabel("position (m)");
```

```python
D = kB * T / (m * gamma)
print(D)
```

```python
MSD_theory = 2 * D * times
```

```python
plt.plot(times, MSD, label="simulation")
plt.plot(times, MSD_theory, label="theory")
plt.xlabel("time (s)")
plt.ylabel(r"$\mathrm{MSD}(t)$ (m$^2$)")
plt.legend();
```

```python
from scipy.optimize import curve_fit
def msd(x, m, c):
  return m*x+c

parameters, covariance = curve_fit(msd, times, MSD)
fit_m = parameters[0]
fit_c = parameters[1]
print(fit_m)
print(fit_c)
fit_y = msd(times, fit_m, fit_c)
plt.plot(times,MSD , linestyle ='-',label='data', lw = 2, color = 'black')
plt.plot(times, fit_y, '-', label='fit', color = 'red')
plt.plot(times, MSD_theory, label="theory")
plt.title('means squared displacement', fontsize = 14)
plt.ylabel(r'$\left<x^2\right>$', fontsize = 14)
plt.xlabel('time', fontsize = 14)
plt.legend()
```

```python
from scipy.optimize import curve_fit
def msd(t, D):
  return 2*D*t

parameters, covariance = curve_fit(msd, times, MSD)
D_fit = parameters[0]

```
